Q3 (a) Create a tensor X of shape 2 × 2, filled with random floats, and
requires gradients.
(b) Define a function f (X) = P X2, and compute the gradient of f with
respect to X using backpropagation.
(c) Print the gradient of X.

```python
# Create tensor with gradient tracking enabled
X = torch.randn(2, 2, requires_grad=True)

print("X:")
print(X)
```

```python
# Define function
f = torch.sum(X**2)

# Backpropagation
f.backward()
```

```python
print("Gradient of X:")
print(X.grad)
```

Q5 (a) Create a simple linear model with a single layer using PyTorch’s
neural network module.
(b) Initialise a stochastic gradient descent (SGD) optimiser for the model’s
parameters with a learning rate of 0.01.
(c) Create a dummy target tensor and compute the mean squared error
loss between the model’s output and the target.
(d) Use the optimiser to update the model’s parameters based on the
computed loss.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Simple linear model: 3 inputs → 1 output
model = nn.Linear(in_features=3, out_features=1)

print(model)
```

```python
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# Dummy input (5 samples, 3 features)
X = torch.randn(5, 3)

# Dummy target (5 samples, 1 output)
target = torch.randn(5, 1)

# Forward pass
output = model(X)

# Mean Squared Error loss
criterion = nn.MSELoss()
loss = criterion(output, target)

print("Loss:", loss.item())
```

```python
# Backprop + update
optimizer.zero_grad()
loss.backward()
optimizer.step()

print("Updated parameters:")
for param in model.parameters():
    print(param)
```
