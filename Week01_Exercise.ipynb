### Q1 ###
Write a Python program that asks the user for a number and prints the
sum of all numbers from 1 to the input number using a loop. Include error
handling to ensure the input is a positive integer

```python
y = 0

while True:
    x = int(input("Please enter a positive integer: "))
    if x > 0:
        break
    print("That is not a positive integer. Try again.")

for i in range(1, x + 1):
    y += i

z = x * (x + 1) // 2

print(y)
print(z)
```

### Q2 ###
Create a Python class named Rectangle with two attributes, length and
width. Write a method within the class that calculates the area of the
rectangle. Then, write a function outside of the class that takes an instance
of Rectangle and prints the area

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

x = int(input("Please enter length: "))
y = int(input("Please enter width: "))

rect = Rectangle(x, y)
print(rect.area())
```

### Q3 ###
Using the Numpy package, generate a 10 × 10 array of random integers
between 1 and 100. Write a function that replaces all elements in the array
that are greater than 50 with the value −1. Demonstrate this functionality
with an example.

```python
import numpy as np

arr = np.random.randint(1, 101, size=(10, 10))
print("Original array:")
print(arr)
def replace_gt_50(a):
    a = a.copy()        # avoid modifying original array
    a[a > 50] = -1
    return a
new_arr = replace_gt_50(arr)
print("Modified array:")
print(new_arr)
```

### Q4 ###
Create a 1D Numpy array of 100 evenly spaced numbers between 0 and
2π. Calculate the sine and cosine of these numbers using Numpy’s broad-
casting feature. Use Matplotlib to plot both the sine and cosine functions
on the same graph, with appropriate labels and a legend.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Create a 1D array of 100 evenly spaced values between 0 and 2π
x = np.linspace(0, 2 * np.pi, 100)

# 2️⃣ Calculate sine and cosine (NumPy broadcasting)
sin_x = np.sin(x)
cos_x = np.cos(x)

# 3️⃣ Plot both on the same graph
plt.figure()
plt.plot(x, sin_x, label="sin(x)")
plt.plot(x, cos_x, label="cos(x)")
plt.xlabel("x (radians)")
plt.ylabel("Value")
plt.title("Sine and Cosine Functions")
plt.legend()
plt.show()
```

### Q5 ###
Load an image using OpenCV and convert it to grayscale. Then, using
the scikit-learn library, divide the image into a grid of 4 × 4 pixel regions
and compute the average colour of each region. Display the resulting grid
of average colours as an image. Make sure to include comments explaining
each step of your process.

```python
import os
if not os.path.exists('images.zip'):
  !wget --no-check-certificate https://empslocal.ex.ac.uk/people/staff/ad735/ECMM426/images.zip
  !unzip -q images.zip
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Load the image using OpenCV
img = cv2.imread("images/books.jpg")

# OpenCV loads images in BGR, convert to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2️⃣ Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3️⃣ Ensure image dimensions are divisible by 4
h, w = gray.shape
gray = gray[:h - h % 4, :w - w % 4]

# Number of patches along height and width
n_patches_y = gray.shape[0] // 4
n_patches_x = gray.shape[1] // 4

# 4️⃣ Reshape the grayscale image into non-overlapping 4x4 blocks
# and 5️⃣ Compute the average intensity of each 4x4 region
avg_image = gray.reshape(n_patches_y, 4, n_patches_x, 4).mean(axis=(1, 3))

# 6️⃣ Display the resulting grid of average values
plt.imshow(avg_image, cmap='gray')
plt.title("Average Colour per 4×4 Pixel Region")
plt.axis('off')
plt.show()
```
